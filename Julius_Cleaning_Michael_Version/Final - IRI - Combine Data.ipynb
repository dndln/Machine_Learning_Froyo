# Panelist Files

```python
filteredfiles = [x for x in onlyfiles if 'PANEL' in x]
filteredfiles
```

```python
#check if headers differ accross files
listheaders = list()
rowcounts = list()
i = 0
print 'Rows:\t\t', 'Filename:\t\t', 'file count:'
for x in filteredfiles:
    i += 1
    header = !head -1 Data/$x
    rc = !wc -l Data/$x
    # Row Count, FileName, FileCount, 
    rowcounts.append((int(rc[0].split()[0]), x, header[0].find('\t'),header[0].find(',')))
    print rc[0].split()[0], '\t', x, '\t', i
    listheaders.append(header[0])

headers = set(listheaders)
print '\nDistinct Headers found:', len(set(listheaders)), '\n'
for h in headers:
    print [h]

print '\nTotal rows:', sum([x[0] for x in rowcounts]) - len(rowcounts)
print 'Total files:', len(rowcounts)
```

### Load all files into a single dataframe'

```python
files = list()
for c, f, tab, comma in rowcounts:
    if tab >=0:
        df = pd.read_csv(mypath + f, sep='\t')
    if comma >=0:
        df = pd.read_csv(mypath + f, sep=',')
    if tab + comma <= -2:
        df = pd.read_csv(mypath + f, sep=' ',index_col=False)
    df['file'] = f
    if c-1 != len(df):
        print f, 'dataframe row count mismatch!'
    else:
        print df.shape
    files.append(df)

panel = pd.concat(files)

#check if total row count is correct
if sum([x[0] for x in rowcounts]) - len(rowcounts) != len(panel):
    print 'dataframe row count mismatch! source, dataframe:', sum([x[0] for x in rowcounts]) - len(rowcounts), len(panel)
else:
    print 'Rows in Dataframe:',  len(panel)
print panel.dtypes
```
